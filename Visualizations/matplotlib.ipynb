```python
import matplotlib.pyplot as plt
```

## 1. Line Plot - Continuous values

```python
x = [1,2,3,4,5]
y = [0, 1.5, 2, 2.8, 6]
```

```python
plt.plot(x, y)
```

```python
# Customization

plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("Basic Plot")
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='--', marker = 'o')
```

```python
y1 = [1, 4, 9, 16, 25]

plt.figure(figsize=(9,5))               # Figure size 900x500

plt.subplot(2, 2, 1)                    # subplot(rows, columns, position)
plt.plot(x, y, color='red')
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y1, color='green')
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(y, x, color='green', linestyle='--')
plt.title('Plot 3')


```

## 2. Bar Plot - Categorical Values

```python
items = ['A', 'B', 'C', 'D']
values = [2, 4, 1, 5]

plt.bar(items, values, color='orange')
```

## 3. Histogram - for distribution of data

```python
data = [1, 1, 2, 1, 3, 1, 2,3, 2, 3, 5, 4, 5, 5,5, 1, 3, 1, 4]

plt.hist(data ,bins=5, color='orange', edgecolor = 'black')
```

## 4. Scatter Plot - relation between fields

```python
plt.scatter(x, y, marker = 'o')
```

```python
items = ['A', 'B', 'C','D']
occupy = [20, 30, 40, 10]
colors = ['red', 'violet', 'orange', 'lightblue']
explode = [0,0,0,0.2]

plt.pie(occupy, labels = items, colors = colors, explode=explode, shadow = True, autopct = '%1.1f%%')
plt.show()
```

## 5. Sample Visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.isnull().sum()
```

```python
# Total sales of each product
new_table_1 = df.groupby('Product')['Total_Sales'].sum().reset_index()
new_table_1
```

```python
plt.figure(figsize=(14,7))
plt.bar(new_table_1['Product'], new_table_1['Total_Sales'], color='teal')
plt.title('Product and it Sale')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```
