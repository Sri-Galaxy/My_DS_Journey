## Iterators

For Efficient looping and memory management as it uses lazy loading. Allows to access elements sequentially without exposing underlying structure.

```python
arr = [12,3,5,65,5]

iterator = iter(arr)
```

```python
while True:
    try:
        print(next(iterator))
    except StopIteration:
        break
```

## Generators

A simple way to create iterators. They use yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store in memory.

```python
def gen_text(file):
    with open(file, 'r') as f:
        for line in f:
            yield line

for i in gen_text('input.txt'):
    print(i.strip())
```
