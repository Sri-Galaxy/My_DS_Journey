## Lambda Expressions

```python
# -------------------   SYNTAX    :: var = lambda x1, x2, x3, ... : expression  ------------------

# Simple Implementation...
add = lambda x, y: x + y
print(add(2,5))

# If-Else in lambda Expression...
is_even_odd = lambda n: "Even" if (n % 2 == 0) else "Odd"
print(is_even_odd(7))

# If-Else ladder in lambda expression...
category = lambda n: "Positive" if (n > 0) else ("Negative" if (n < 0) else "Zero")
print(category(0))
```

## Map

```python
# SYNTAX    ::  map(function, iterable)


# def square(n):
#     return n*n

arr = [1,2,3,4,5]
print(list(map(lambda x: x*x, arr)))    # We can also pass actual function like, map(square, arr)

print(list(map(lambda x: x%2 == 0, arr)))

arr2 = [1,2,3]
print(list(map(lambda x1, x2: x1+x2, arr, arr2)))

arr3 = ['apple', 'BANANA', 'Orange']
print(list(map(str.upper, arr3)))     # we can also use, map(lambda x: x.upper(), arr3)
```

## Filter

```python
print(list(filter(lambda x: x%2==0, arr)))

print(list(filter(lambda x: x > 3 and x%2==0, arr)))

details = [
    {'name':'Srinath', 'age': 21},
    {'name':'Rizu', 'age': 20},
    {'name': 'Rohi', 'age': 25}
]

print(list(filter(lambda x: x['age'] > 21, details)))
```
