```python
# Checking for missing values
df.isnull().sum()
```

```python
# Checking for duplicate records
df.duplicated()         # Returns boolean Series denoting duplicate rows
df.duplicated().sum()   # Total number of duplicate rows
```

```python
# Removing duplicate records
df.drop_duplicates(inplace=True)

# df = df[~df.duplicated()]  Keep only non-duplicate rows
df.duplicated().sum()
```

```python
df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
df.quality.value_counts()   # we can cleary see it is a imbalanced dataset 
```

```python
plt.bar(df['quality'].unique(), df['quality'].value_counts())
plt.xlabel('wine quality')
plt.ylabel('count')
plt.show()
```
