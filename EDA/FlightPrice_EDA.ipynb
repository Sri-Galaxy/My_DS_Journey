```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date_of_Journey'].apply(lambda x: int(x.split('/')[0]))
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
```

```python
df.drop('Date_of_Journey', inplace = True, axis=1)
df.info()
```

```python
df.head(2)
```

```python
df["Arrival_Hours"] = df['Arrival_Time'].apply(lambda x: int(x.split(':')[0]))
df["Arrival_Minutes"] = df['Arrival_Time'].apply(lambda x: int(x.split(':')[1][:2]))

df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df["Dep_Hours"] = df['Dep_Time'].apply(lambda x: int(x.split(':')[0]))
df["Dep_Minutes"] = df['Dep_Time'].apply(lambda x: int(x.split(':')[1][:2]))

df.drop("Dep_Time", inplace=True, axis=1)
```

```python
df.head(2)
```

```python
df.drop(['Route', 'Additional_Info'], inplace=True, axis=1)
```

```python
print(df['Total_Stops'].unique())  # We have a nan value
print(df["Total_Stops"].isnull().sum())
```

```python
df["Total_Stops"].mode()
```

```python
df['Total_Stops'] = df["Total_Stops"].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, np.nan: 1})
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

data = encoder.fit_transform(df[['Airline', 'Source', "Destination"]]).toarray()
```

```python
df_encoded = pd.DataFrame(data, columns=encoder.get_feature_names_out(['Airline', 'Source', "Destination"]))
df_encoded.head()
```

```python
df = pd.concat([df, df_encoded], axis=1)

df.drop(['Airline', 'Source', "Destination"], axis=1, inplace=True)
df.head()
```

```python
# extract hours and minutes with a regex
df[['h','m']] = df['Duration'].str.extract(r'(?:(\d+)\s*h)?\s*(?:(\d+)\s*m)?')

# replace NaN with 0 and convert to ints
df['h'] = df['h'].fillna(0).astype(int)
df['m'] = df['m'].fillna(0).astype(int)

df.drop('Duration', axis=1, inplace=True)
```
