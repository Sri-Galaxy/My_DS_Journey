```python
df_filled = df.fillna(0)
df_filled
```

```python
df['sw_filled'] = df['sepal_width'].fillna(df['sepal_width'].median()).astype(int)
df
```

```python
df = df.rename(columns = {'species': 'varieties'})

df.drop('sw_filled', inplace=True, axis=1)
df
```

```python
# Aggregation
# SYNTAX: "df.groupby([column1, column2, ...])[target column (OPTIONAL)].agg([func1, func2, func3, ...])"

group = df.groupby('varieties').count()
group

grouped = df.groupby(['varieties'])['sepal_length'].agg(['mean', 'sum', 'count'])
grouped
```

```python
# Merging

data1 = {
    'Name' : ['a', 'b', 'c'],
    'Age': [12, 13, 16]
}

data2 = {
    'Name' : ['a', 'b', 'd'],
    'Marks': [98, 97, 19]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df1
```

```python
df2
```

```python
merge = pd.merge(df1, df2, on = 'Name', how='inner')
merge
```

```python
merge = pd.merge(df1, df2, on = 'Name', how='left')
merge
```

```python
merge = pd.merge(df1, df2, on = 'Name', how='right')
merge
```

```python
merge = pd.merge(df1, df2, on = 'Name', how='outer')
merge
```
